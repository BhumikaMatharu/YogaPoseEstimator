```python
import pandas as pd
input_data = pd.read_csv('output_joints3d.csv', header=None, sep=' ')
print(input_data.shape)
input_data = input_data.round(3)
input_data.drop_duplicates(inplace=True)
input_data.shape
```

```python
X, y = input_data.iloc[:, :-1], input_data.iloc[:, [-1]]
X.shape
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
dummy_y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, dummy_y, random_state = 0, stratify=y, test_size=0.2) 
```

```python
y_train[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.utils import to_categorical
```

```python
model = Sequential()
model.add(Dense(8, input_dim = X_train.shape[1] , activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(11, activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy' , optimizer = 'adam' , metrics = ['accuracy'] )

model.fit(X_train, y_train, epochs = 3, batch_size = 2)
```

```python
scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
scores1 = model.evaluate(X_train, y_train)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores1[1]*100))
```

```python
model.summary()
```

```python
model_test = Sequential()
model_test.add(Dense(8, input_dim=X_train.shape[1], activation='relu'))
model_test.add(Dense(11, activation='softmax'))
# Compile model
model_test.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_test.fit(X_train, y_train, epochs = 10, batch_size = 2)
```

```python
scores_test = model_test.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model_test.metrics_names[1], scores_test[1]*100))
scores_test_1 = model_test.evaluate(X_train, y_train)
print("\n%s: %.2f%%" % (model_test.metrics_names[1], scores_test_1[1]*100))
```
